```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

%reload_ext autoreload
%autoreload 2
```

```python
x_1 = np.linspace(0,5,10)
y_2 = x_1**2
plt.plot(x_1,y_2)
plt.title('Squared title')
plt.xlabel('Numbers')
plt.ylabel('Number Square')
```

```python
plt.subplot(1,2,1)
plt.plot(x_1,y_2,'r')
plt.subplot(1,2,2)
plt.plot(x_1,y_2,'g')
```

```python
fig1 = plt.figure(figsize=(5,4), dpi=100)
axes_1 = fig1.add_axes([0.1,0.1,0.8,0.8])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Square')
axes_1.set_title('Days Square Chart')
axes_1.plot(x_1,y_2, label='x/x^2')
axes_1.plot(y_2,x_1, label='x/x^2')
axes_1.legend(loc=0)


axes_2 = fig1.add_axes([0.45,0.45,0.4,0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Square')
axes_2.set_title('Days Square Chart')
axes_2.plot(x_1,y_2, 'r')

axes_2.text(0,40, 'Hello')
```

```python
fig_2, axes_2 = plt.subplots(figsize=(8,4), nrows=1, ncols=3)
plt.tight_layout()
axes_2[1].set_title('Plot 2')
axes_2[1].plot(x_1,y_2)
```

```python
fig_3 = plt.figure(figsize=(6,4))
axes_3 = fig_3.add_axes([0,0,1,1])
axes_3.plot(x_1,y_2, color="#ff0000", alpha=.75, lw=2, ls='-.', marker='o',
           markersize=7, markerfacecolor='b', markeredgecolor='y', markeredgewidth=4)

# axes_3.set_ylim([0,8]) Set length of y axis

axes_3.grid(True, color='0.6', dashes=(5,2,1,2))
axes_3.set_facecolor('#FAEBD7')
```

```python
df = pd.read_csv('icecreamsales.csv')
df = df.sort_values(by='Temperature')

np_array = df.values
x_2 = np_array[:,0]
y_2 = np_array[:,1]
fig4 = plt.figure(figsize = (6,4))
axes_4 = fig4.add_axes([0,0,1,1])

axes_4.set_xlabel('Temperature')
axes_4.set_ylabel('IceCream Sales')
axes_4.set_title('Ice_cream - Temperature')
axes_4.plot(x_2,y_2)

axes_4.annotate('Good Month', xy=(83,536), xytext=(60, 520),
               arrowprops=dict(facecolor='black', shrink=0.05))
```
